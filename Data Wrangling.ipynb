# Label Encoding

# Method 1

```python
from sklearn.preprocessing import LabelEncoder
from collections import Counter as count
le=LabelEncoder()
print('Before label Encoding: ',count(df['Country']))
df['Country']=le.fit_transform(df['Country'])
print('After label Encoding: ',count(df['Country']))
```

# Method 2

```python
print('Before Replace: ',count(df['Purchased']))
df['Purchased']=df['Purchased'].replace(['No','Yes'],[0,1])
print('After Replace: ',count(df['Purchased']))
```

# One Hot Encoding

```python
from sklearn.compose import ColumnTransformer #it is used for performing both OHE & LE in single step
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer([('on',OneHotEncoder(),[0])],remainder='passthrough')
df=ct.fit_transform(df)
```

```python
df.shape
```

```python
df
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
df=sc.fit_transform(df)
```

# Train,Test & Split

```python
#Independent Features
x=pd.DataFrame(df[:,0:5],columns=['France','Germany','Spain','Age','Salary'])
x
```

```python
#Dependent Variable
y=pd.DataFrame(df[:,5:],columns=['Purchased'])
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```
